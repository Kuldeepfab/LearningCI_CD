```python
# Creating Txt File consisting X Y and E in forward and revered direction.
import math
import numpy as np
from math import *
radius=int(input("Enter the Radius:"))
Angle=int(input("Enter the Angle in Degree: "))
step=Angle
Angle_radian=math.radians(Angle)
theta = np.linspace( 0 , Angle_radian , step )
x = radius * np.cos( theta ) 
y = radius * np.sin( theta )
alpha=0
f = open("coordinates.txt", "a")
f.write("\n")
f.write(";START")
f.close
for i in range(0,step):
    if alpha<1: # for second coorinates to add the distance side of the cooridnates
        f = open("coordinates.txt", "a")
        f.write("\n")
        f.write("G1 X {:.2f} Y {:.2f}".format(x[i],y[i]))
        f.close
        alpha+=1
    else:
        distance=sqrt((x[i]-x[i-1])**2 + (y[i]-y[i-1])**2) # calculating distance between two coordinates (x1,y1) and (x2,y2)
        f = open("coordinates.txt", "a")
        f.write("\n")
        f.write("G1 X {:.2f} Y {:.2f} E {:.2f}".format(x[i],y[i],distance))
        f.close    
f = open("coordinates.txt", "a")
f.write("\n")
f.write(";REVERSE")
f.close
beta=0
for j in range(step-1,-1,-1): # for revere points
    if beta<1: 
        f = open("coordinates.txt", "a")
        f.write("\n")    
        f.write("G1 X {:.2f} Y {:.2f}".format(x[j],y[j]))
        f.close
        beta+=1
    else:
        distance=sqrt((x[j]-x[j+1])**2 + (y[j]-y[j+1])**2)
        f = open("coordinates.txt", "a")
        f.write("\n") 
        f.write("G1 X {:.2f} Y {:.2f} E {:.2f}".format(x[j],y[j],distance)) 
        f.close
f = open("coordinates.txt", "a")
f.write("\n")
f.close()  
#If want to see the circle ( for visualization) 
#figure, axes = plt.subplots( 1 ) 
#axes.plot( x, y ) 
#axes.set_aspect( 1 ) 

# Sucessfully added in forward and reversed list.
```
