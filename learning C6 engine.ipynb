```python
# running all reset centre ,scalling, rotation,translation on the mesh.( this is not the part of Slice_file function)
##  to use CreateSection function and understand its role.
import trimesh
import math
your_mesh = trimesh.load_mesh("/Users/peedluek/Desktop/support.stl")
def resetCentroid(mesh):
    # Calculate 3D centroid from bounds
    box_center = [round((float(mesh.bounds[0][0]+mesh.bounds[1][0])/2),2),
            round((float(mesh.bounds[0][1]+mesh.bounds[1][1])/2),2),
            round((float(mesh.bounds[0][2]+mesh.bounds[1][2])/2),2)]

    return translateSTL(mesh, - box_center[0], - box_center[1], - box_center[2])

def translateSTL(mesh, x, y, z):
    translate_matrix = np.eye(4)
    translate_matrix[0, 3] = x
    translate_matrix[1, 3] = y
    translate_matrix[2, 3] = z
    return mesh.apply_transform(translate_matrix)

def scaleSTL(mesh ,x, y, z):
    scale_matrix = np.eye(4)
    xScale = x
    yScale = y
    zScale = z
    scale_matrix[0, 0] *= xScale
    scale_matrix[1, 1] *= yScale
    scale_matrix[2, 2] *= zScale
    return mesh.apply_transform(scale_matrix)

def rotateSTL(mesh,x,y,z):
    xRotation = x
    yRotation = y
    zRotation = z
    rot_matrix = trimesh.transformations.euler_matrix(math.radians(
        xRotation), math.radians(yRotation), math.radians(zRotation), 'rxyz')
    return mesh.apply_transform(rot_matrix)

your_mesh=resetCentroid(your_mesh)
your_mesh=scaleSTL(your_mesh,1,1,1)
your_mesh=rotateSTL(your_mesh,0,0,0)
your_mesh=translateSTL(your_mesh,160,160,12)


#logTransformValues(variables.perFileInput[section], zPos)
#filesToCombine.append(your_mesh)
your_mesh.show()
```

```python
# this function is used to create a box with mutiple boundaries, depend upon lineGap, in both the horizonal and vertiical direction
def lineGeneration(stats, lineGap=0.4606194490192345):
    # Bounding box for STL model
    # Step 1 : Find bounding box for the polygon
    # Step 2 : Draw infill lines for the entire STL

    minX = stats['extents']['x']['lower']    #145.00001525878906
    maxX = stats['extents']['x']['upper']     #175.0
    minY = stats['extents']['y']['lower']     #155.0
    maxY = stats['extents']['y']['upper']     #165.0
    xRange = maxX-minX
    yRange = maxY-minY
    maxRange = max(xRange, yRange)
    minX -= maxRange  # ?
    maxX += maxRange  # ?
    minY -= maxRange  # ?
    maxY += maxRange  # ?

    lineCoord1 = np.arange(minY, maxY, lineGap)
    lineCoord2 = np.arange(minX, maxX, lineGap)
    infillLines1 = [] # for vertical boundaries of box
    infillLines2 = [] # for horizontal boundaries of box
    for cord in lineCoord1:
        line = LineString([(minX, cord), (maxX, cord)])
        infillLines1.append(line)
    for cord in lineCoord2:
        line = LineString([(cord, minY), (cord, maxY)])
        infillLines2.append(line)
    return infillLines1, infillLines2

from shapely.geometry import Point, LineString





topBottomInfillLines1, topBottomInfillLines2 = lineGeneration(stats)
#Understanding linegeneration function,
printpoly2(topBottomInfillLines1[:])
#Understanding linegeneration function,
printpoly2(topBottomInfillLines2[:])
```

```python
def infillPathMultiproc(loopStartLayer, totalLayers, layerLoop, stats, infill_type, pathSpace, perimeterCount, globalCentroid, queueIndex=None):
    layerInfillPath = {}
    #perimeterInfillPathForGCode = {}
    innerPerimeterInfillPathForGCode = {}
    outerPerimeterInfillPathForGCode = {}
    #fiberWallInfillPathForGCode = {}
    outerFiberWallInfillPathForGCode= {}
    innerFiberWallInfillPathForGCode = {}
    layerSlicePolygons = []
    infillAngleList = []
    lineGapList = []
    polyOrgOptimizedList = []
    polyRevOptimizedList = []
    polygonBodiesList = []
    outerPerimeterPolygonList = []
    innerPerimeterPolygonList = []
    multiPolyList = []
    multiPolyRevList = []
    multiPolySmallList = []
    multiPolySmallRevList = []
    finalPolygonList = []
    endPolyUnion = {}
    endPolyUnionRev = {}
    endPolyUnionSmall = {}
    endPolyUnionSmallRev = {}
    outerPolygonForBrim = []  # Empty list
    multiProcLogs = ""
    wallProcessShouldExit = False
    for layerIndex, layerNo in enumerate(range(loopStartLayer, loopStartLayer+len(layerLoop))): #0 to 134 run in our case
        # variables.logs += f"========= Layer No. = {layerNo} =========" +"\n"
        print("Layer : ",layerNo)
        endPolyUnion[layerNo] = [] # creating empty list inside dictationary
        endPolyUnionRev[layerNo] = []
        endPolyUnionSmall[layerNo] = []
        endPolyUnionSmallRev[layerNo] = []
        layerInfillPath[layerNo] = []
        outerPerimeterInfillPathForGCode[layerNo] = []
        innerPerimeterInfillPathForGCode[layerNo] = []
        outerFiberWallInfillPathForGCode[layerNo] = []
        innerFiberWallInfillPathForGCode[layerNo] = []
        multiPolyList.append([])
        multiPolyRevList.append([])
        multiPolySmallList.append([])
        multiPolySmallRevList.append([])
        polyOrgOptimizedList.append([])
        polyRevOptimizedList.append([])
        infillAngle = float(variables.config['infill']['INFILL_ANGLE']) if (
            infill_type != 'concentric') else 0  # check if infill is conecentric then infill is Zero ,else frontend, but in our case it is rectilinear so infill angle is 45 , according to config.ini 
        if len(layerLoop[layerIndex]) == 0: ## Check empty loop of empty then add polygon
            finalPolygonList.append([Polygon([])])
        if len(layerLoop[layerIndex]) != 0:   ## if not empty then turn the infill True
            # variables.selfPrint("Layer Number : ", layerNo)
            isInfillLayer=True   
            # Infill Density for bottom & Top layer will be different from middle layers
            if(layerNo < int(variables.config['infill']['BOTTOM_LAYER_COUNT'])   #if it is True top and bottom layer calulation begins  and make infill FALSE, becoz it will be filled solid: else we go for middle layer
               or layerNo >= totalLayers-int(variables.config['infill']['TOP_LAYER_COUNT'])):
                lineGap = utils.lineGapCalculation(1.0)   # in case of top and bottom layer, we do lineGAPcalculation again
                isInfillLayer=False  #
            else:
                if(variables.config['fiber'].getboolean('ALTERNATE_FIBER_LAYER') == True): # if alternate fiber is TRUE
                    isInfillLayer=False if(layerNo%2==0) else True
                lineGap = utils.lineGapCalculation()
                """ if(variables.config['fiber'].getboolean('USE_FIBER') == True):
                    pathSpace=float(variables.config['fiber']['FIBER_EXTRUSION_WIDTH'])  """
                # No infill-angle for honeycomb infill at the moment - TODO
            if(infill_type == 'honeycomb'): # in our default case it is rectilinear
                infillAngle = 0              # resetting the infill angle to Zero, earlier it was 45 ??
            #variables.logs += f"pathSpace = {pathSpace}" +"\n"
            infillAngleList.append(infillAngle)
            lineGapList.append(lineGap)
            ####========================================####
            # This functionality was merged in slicing step using polygonize
            # If number of points is less than 3, polygon formation is not possible.
            # So we need to remove thoose loops and directly add the points

            # Check all the polygon formed from the coordinates of layerLoop lists is valid or not.
            # If not valid make it valid  and insert the corrected polygon
            ####========================================####

            layerSlicePolygons.append(layerLoop[layerIndex])
            # Segregating inner and outer polygon
            # And offsetting outer and inner perimeters by pathspace/2
            """For each layerloop we are sepearting polygon bodies,Outerpolygon,OutermultipolygonCounts"""
            polygonBodies, outerPolygons, outerMultiPolygonCount = utils.separateOuterInner(
                layerLoop[layerIndex], pathSpace)
            
            polygonBodiesList.append(polygonBodies)
            if(layerNo == 0):  # For brim generation
                outerPolygonForBrim = outerPolygons

            ####========================================####
            # This functionality was merged in separateOuterInner method
            #
            # Arrange all outer and inner loops in a list
            # where in each element - at index 0 it contains outerpolygon and innerpolygon in all other index
            ####========================================####
            
            
            """---------kd Added comments------
            # for each polygonBodies we calculate finalpolygons,OuterperimeterCoords,innerperimetercoordinates
            # after calculating for each polygonbody we append the outer and innerPerimeterpolygon into main outerPerimeterPolygonList."""

            finalPolygon = []
            # polyPerimeter={}
            count = 0
            outerPerimeterPolygon = {}
            innerPerimeterPolygon = {}
            fiberOuterPerimeterPolygon = {}
            fiberInnerPerimeterPolygon = {}
            # For each body (or each outer polygon)
            for polyNum1 in range(len(polygonBodies)):
                outerPerimeterPolygon[polyNum1] = []
                innerPerimeterPolygon[polyNum1] = []
                fiberOuterPerimeterPolygon[polyNum1]=[]
                fiberInnerPerimeterPolygon[polyNum1]=[]
                if(variables.config['fiber'].getboolean('USE_FIBER')==True and isInfillLayer):
                    infillPolygons, outer, inner,fiberOuter,fiberInner, exit = walls.perimeterGeneration(polygonBodies[polyNum1],pathSpace, perimeterCount,isInfillLayer)
                    fiberOuterPerimeterPolygon[polyNum1].append(fiberOuter)
                    fiberInnerPerimeterPolygon[polyNum1].append(fiberInner)
                else:
                    infillPolygons, outer, inner, exit = walls.perimeterGeneration(polygonBodies[polyNum1],pathSpace, perimeterCount,isInfillLayer)
                wallProcessShouldExit = wallProcessShouldExit or exit

                outerPerimeterPolygon[polyNum1].append(outer)
                innerPerimeterPolygon[polyNum1].append(inner)
                    
                try: 
                    for ply in infillPolygons:
                        finalPolygon.append(ply)
                except:
                    finalPolygon.append(infillPolygons)

            outerPerimeterPolygonList.append(outerPerimeterPolygon) # finally the main list of outerPerimeterPolygonlist is updated .
            innerPerimeterPolygonList.append(innerPerimeterPolygon) # finally the main list of innerPerimeterPolygonlist is updated .




            """KD_step1: check each polygon of finalpolygon list: if empty add the index number of that polygon into a seperate list, later that will be deleted from finalpolygon list
            if it is not empty ,then calculate the distance between centroid of polygon  and origin and save into a list"""
            polyDistFromOrigin = []  # This stores the distance of the object from the origin
            delFinalPolyIdx = []
            # Distance calculation from "origin" so timing can be optimized
            for polyNum in range(len(finalPolygon)):
                # if final polygon is empty we will delete it
                if(finalPolygon[polyNum].is_empty):
                    delFinalPolyIdx.append(polyNum)
                else:
                    centroid = Point(finalPolygon[polyNum].centroid.coords)
                    polyDistFromOrigin.append(centroid.distance(Point(0, 0)))

            # Deleting all the empty polygon
            # Should be done before polygon reduction above - TODO
            delFinalPolyIdx.reverse()
            for idx in delFinalPolyIdx:
                del finalPolygon[idx]
                
                
            """Kd_   Sorting of finalpolygon list with respect to the distance from origin"""
            # Arrange the polygon with respect to its distance from the origin(Essential in multi object infill)
            finalPolygon = np.array(finalPolygon)
            indexSort = np.argsort(polyDistFromOrigin)
            indexSort = indexSort.tolist()
            finalPolygon = finalPolygon[np.argsort(polyDistFromOrigin)]

            """ Adding perimeter lines"""
            # inserting outerperimeterpolygonlist into the OuterInfillPathForGcode dict on particular index number(i.e layer number)  
            for polyNumi in range(len(outerPerimeterPolygon)):
                for polyNumj in range(len(outerPerimeterPolygon[polyNumi])):
                    for perimeter in outerPerimeterPolygon[polyNumi][polyNumj]:
                        outerPerimeterInfillPathForGCode[layerNo].append(perimeter)
            # CHECK for knowing the preference of print, either innerwall first or outer first, if inner first then reverse.
            if(variables.config['printParameters'].getboolean('PRINT_INNER_WALL_FIRST')==True):
                outerPerimeterInfillPathForGCode[layerNo].reverse()
            
            # inserting innerperimeterpolygonlist into the innerInfillPathForGcode dict.
            for polyNumi in range(len(innerPerimeterPolygon)):
                for polyNumj in range(len(innerPerimeterPolygon[polyNumi])):
                    for perimeter in innerPerimeterPolygon[polyNumi][polyNumj]:
                        innerPerimeterInfillPathForGCode[layerNo].append(perimeter)
            if(variables.config['printParameters'].getboolean('PRINT_INNER_WALL_FIRST')==True):
                innerPerimeterInfillPathForGCode[layerNo].reverse()
                
            """ Adding fiber perimeter lines"""
            if(variables.config['fiber'].getboolean('USE_FIBER')==True and isInfillLayer):
                
                #inserting fiberOuterPerimeterPolygon into the outerFiberWallInfillPathForGCode dict.
                for polyNumi in range(len(fiberOuterPerimeterPolygon)):
                    for polyNumj in range(len(fiberOuterPerimeterPolygon[polyNumi])):
                        for perimeter in fiberOuterPerimeterPolygon[polyNumi]:
                            outerFiberWallInfillPathForGCode[layerNo].append(perimeter)
                            
                # CHECK for knowing the preference of print, either innerwall first or outer first, if inner first then reverse.
                if(variables.config['printParameters'].getboolean('PRINT_INNER_WALL_FIRST')==True):
                    outerFiberWallInfillPathForGCode[layerNo].reverse()
                for polyNumi in range(len(fiberInnerPerimeterPolygon)):
                    for polyNumj in range(len(fiberInnerPerimeterPolygon[polyNumi])):
                        for perimeter in fiberInnerPerimeterPolygon[polyNumi][polyNumj]:
                            innerFiberWallInfillPathForGCode[layerNo].append(perimeter)
                if(variables.config['printParameters'].getboolean('PRINT_INNER_WALL_FIRST')==True):
                    innerFiberWallInfillPathForGCode[layerNo].reverse()
                
            finalPolygonList.append(finalPolygon)
            
            
            """Step1: Check for empty polygon body in finalpolygon list
            Step2: if not empty , apply polygon ,rotate twice, one at infillangle other at 90 +infill angle (for grid)"""
            # For each body (only infill polygons), current layer
            for polyNum, body in enumerate(finalPolygon):
                # centroid=list(finalPolygon[polyNum].centroid.coords)
                if(finalPolygon[polyNum].is_empty):  # IF polygon is empty
                    # print("Empty : ",layerNo)
                    continue

                """ Step 1: First perform Polygon rotation.
                Step 2: Find Maxima and Minimas of the Rotated Polygon.
                Step 3: Perform the polygon infilling at zero degrees.
                Step 4: Get the polygon infilling lines and superimpose on the initial polygon. """

                # polyTranslated=affinity.translate(finalPolygon[polyNum])
                # Rotating the polygon at the infill angle.

                polyRotated = affinity.rotate(
                    body, infillAngle, globalCentroid)
                # Rotating the polygon at the infill +90 angle for grid
                polyRotatedRev = affinity.rotate(
                    body, (infillAngle+90), globalCentroid)
                xCord = []
                yCord = []
                # print(polyRotated)
                
                """Step1:After rotation obtain x outercoordinate and y outercoordinate and append them into xCord ,yCord list
                   Step2:Similarly contain x inner coodinate and y innerCoordinate and append them into list  """
                xOuterCord, yOuterCord = polyRotated.exterior.coords.xy
                xCord.extend(xOuterCord)
                yCord.extend(yOuterCord)
                for interior in polyRotated.interiors:
                    xInnerCord, yInnerCord = interior.coords.xy
                    xCord.extend(xInnerCord)
                    yCord.extend(yInnerCord)

                # To produce grid lines
                """Similary Obtain x,y inner and outercoordinates for reversed polygon( GRID)
                   Because in case of grid we need reverse polygon"""
                   
                xCordRev = []
                yCordRev = []
                xOuterCord, yOuterCord = polyRotatedRev.exterior.coords.xy
                xCordRev.extend(xOuterCord)
                yCordRev.extend(yOuterCord)
                for interior in polyRotatedRev.interiors:
                    xInnerCord, yInnerCord = interior.coords.xy
                    xCordRev.extend(xInnerCord)
                    yCordRev.extend(yInnerCord)

                """If the infilltype is honeycomb then the LineCords LineCordsrev
                 , segmentsLines and segmentlinesreversed will calculate though different method,"""
                 
                if(infill_type == 'honeycomb'):
                    honeyCombSize = utils.honeycombSizeCalculation()    # ??????? functino unknown for me.
                    lineCords = honeycomb.honeyCombComplexPolyZoneLines(
                        polyRotated, honeyCombSize, xCord, yCord, stats)
                    segmentLines = LineString(lineCords)
                    lineCordsRev = honeycomb.honeyCombComplexPolyZoneLines(
                        polyRotatedRev, honeyCombSize, xCordRev, yCordRev, stats)
                    segmentLinesRev = LineString(lineCordsRev)
                else:
                    """This function(complexpolyzonesLines) returns the list of lines(called as segmentlines) which will perfrom polygon slicing i.e. divide the polygon in segments.
                    These segments. obtain lineCoords, that used to create the segements through the polygon"""
                    # Slice lines for polygon rotated at infill angle   ??????????????? function  uknow for me
                    lineCords, _ = polySegLines.complexPolyZoneLines(
                        xCord, yCord, polyRotated.bounds)
                    segmentLines = LineString(lineCords)

                    # Slice lines for polygon rotated at infill+90 angle
                    lineCordsRev, _ = polySegLines.complexPolyZoneLines(
                        xCordRev, yCordRev, polyRotatedRev.bounds)

                    segmentLinesRev = LineString(lineCordsRev)
                """ For shapely, the steps for split in this case were as follows:
                1. Take the polygon and get the exterior boundary of polygon as a line
                2. Take the sliceline(i.e segmentLines) and the polygon exterior and union the two together (this will catch all those intersecting points)
                3. Use the polgyonize method to create polygons from the negative space in the union-ed lines
                4. Iterate through all the resulting polygons and keep only the ones that have a point inside the original polygon """

                unioned = polyRotated.boundary.union(segmentLines)
                unionedRev = polyRotatedRev.boundary.union(
                    segmentLinesRev)  # To create a grid
                # use polygonize geos operator and filter out poygons ouside of original input polygon
                keep_polys = []
                for ply in polygonize(unioned):
                    try:
                        if ply.representative_point().within(polyRotated):
                            keep_polys.append(ply)
                    except:
                        pass
                keep_polysRev = []
                for ply in polygonize(unionedRev):
                    try:
                        if ply.representative_point().within(polyRotatedRev):
                            keep_polysRev.append(ply)
                    except:
                        pass

                #keep_polysRev = [ply for ply in polygonize(unionedRev) if ply.representative_point().within(polyRotatedRev)]
                # remaining polygons are the split polys of original shape
                mp = MultiPolygon(keep_polys)
                mpRev = MultiPolygon(keep_polysRev)

                isNotInfillLayer = (layerNo < int(variables.config['infill']['BOTTOM_LAYER_COUNT'])
                                    or layerNo >= totalLayers-int(variables.config['infill']['TOP_LAYER_COUNT']))

                # Check if polygon area is less than a certain threshold. If less do solid infill.
                multiPoly = []
                multiPolySmall = []
                for poly in mp:
                    # calculating area of the body (full polygon)
                    area = body.area
                    if(isNotInfillLayer):
                        multiPoly.append(poly)  # Is this required ??
                    elif(area <= float(variables.config['infill']['MIN_INFILL_AREA'])): # 100
                        multiPolySmall.append(poly)
                    else:
                        multiPoly.append(poly)

                multiPolyRev = []
                multiPolySmallRev = []
                for poly in mpRev:
                    # calculating area of the body (full polygon)
                    area = body.area
                    if(isNotInfillLayer):
                        multiPolyRev.append(poly)  # Is this required ??
                    elif(area <= float(variables.config['infill']['MIN_INFILL_AREA'])):
                        multiPolySmallRev.append(poly)
                    else:
                        multiPolyRev.append(poly)

                # List of infill polygons in this layer, (optimized for travel time?)
                # polyOrgOptimized = polygonTravelPathOptimization(multiPoly) #TODO- can be optimized
                # polyRevOptimized = polygonTravelPathOptimization(multiPolyRev) #TODO- can be optimized

                multiPolyList[layerIndex].append(multiPoly)
                multiPolyRevList[layerIndex].append(multiPolyRev)
                multiPolySmallList[layerIndex].append(multiPolySmall)
                multiPolySmallRevList[layerIndex].append(multiPolySmallRev)
                # polyOrgOptimizedList[layerIndex].append(polyOrgOptimized)
                polyOrgOptimizedList[layerIndex].append(multiPoly)
                # polyRevOptimizedList[layerIndex].append(polyRevOptimized)
                polyRevOptimizedList[layerIndex].append(multiPolyRev)

                multiProcLogs += f"{layerNo} \t\t|\t\t {len(multiPoly)} \t\t|\t\t {len(multiPolyRev)} \t\t\t\t|\t\t {len(multiPolySmall)} \t\t\t|\t\t\t {len(multiPolySmallRev)} \t\t|\t\t {outerMultiPolygonCount}" + "\n"

    return (lineGapList, polygonBodiesList, finalPolygonList, infillAngleList, outerPerimeterPolygonList, innerPerimeterPolygonList, outerPolygonForBrim, layerSlicePolygons, multiPolyList, multiPolyRevList, multiPolySmallList, multiPolySmallRevList, 
        polyOrgOptimizedList, polyRevOptimizedList, layerInfillPath, outerPerimeterInfillPathForGCode,innerPerimeterInfillPathForGCode,outerFiberWallInfillPathForGCode,innerFiberWallInfillPathForGCode, multiProcLogs, wallProcessShouldExit), queueIndex
```

```python
def perimeterGeneration(polygon,pathSpace,perimeterCount,isInfillLayer):
    "Generates perimeter polygon"
    #polygonPerimeter={}
    outerPerimeterCoords=[]
    outerWallWipeIndices=[] # ? Y
    innerPerimeterCoordsList=[]
    innerPerimeterCoords={}
    outerMostBodyPeri=[]
    innerPolySmall={}
    innerSmallPolyPeriCoords={} # For fiber case if peri length is less than fiber cut dist.
    tempIsInfillLayer=isInfillLayer
    #innerPerimeterPolyCentroid={}
    innerPerimeterArea={}
    infillPolygons=[]
    finalPolyDict={}
    fiberWidth=float(variables.config['fiber']['FIBER_EXTRUSION_WIDTH'])   #2
    fiberOuterPerimeterCoords=[]
    fiberInnerPerimeterCoords=[]
    # INFILL_WALL_OVERLAP 
    # 1 = 1 wall overlap
    # 0 = just touch
    # -1 = 1 wall gap (ideally should be banned)
    if(isInfillLayer and variables.config['fiber'].getboolean('USE_FIBER')==True):
        plastic_wall=int(variables.config['fiber']['PLASTIC_OUTER_WALL_COUNT'])  #1
        fiber_wall=int(variables.config['fiber']['FIBER_WALL_COUNT'])  #1
        perimeterCount=plastic_wall + fiber_wall + 1
        INFILL_WALL_OVERLAP =-0.25 if(variables.config['infill']['INFILL_TYPE']=='rectilinear') else 0 #  Y?
        # complex condition check, which will change fiber wall count
        fiber_wall = checkPerimeterCoalesce(polygon, plastic_wall, fiber_wall, fiberWidth, pathSpace)  # WHAT DOES IT MEAN 
    else:
        plastic_wall=perimeterCount
        fiber_wall=0
        INFILL_WALL_OVERLAP = float(variables.config['printParameters']['INFILL_WALL_OVERLAP_FACTOR'])-1 # Do we not want overlap here ? What happens if user give 0 from front-end ?
    innerPeriCentroidList=[]
    # For each perimeter (outer or inner)
    offset=0
    perimeterOffsetList=getPerimeterOffsetList(pathSpace, perimeterCount, isInfillLayer, plastic_wall, fiber_wall, fiberWidth) # Get a list of offsets based on conditions
    periNum=0
    polygonInnerOuterAfterFiber=None # This will store polygon after fiber walls are done and we transition into plastic walls

    while(periNum<perimeterCount):
        #print(periNum)
        offset+=perimeterOffsetList[periNum]
        if(periNum<plastic_wall):
            polygonInnerOuter=periPolyGeneration(offset,polygon)

            # Define infill polygon
            finalPolygon=polygonDictToPoly(polygonInnerOuter) 

        elif(periNum==plastic_wall):
            polygonInnerOuter=periPolyGeneration(offset,polygon)

            # Length check for outer poly
            outerFlag=False
            for polyNum in range(len(polygonInnerOuter)):
                innerPolySmall[polyNum]=[]
                length=0
                ply=polygonInnerOuter[polyNum][0]
                for indx in range(plastic_wall,plastic_wall+fiber_wall):
                    length+=ply.length
                    ply=ply.buffer(-1*fiberWidth, join_style=1, mitre_limit=15.0)
                if(length<float(variables.config['fiber']['FIBER_CUT_OFFSET'])):
                    isInfillLayer=False
                    for indx in range(periNum,perimeterCount):
                        perimeterOffsetList[indx]=pathSpace
                    plastic_wall=perimeterCount
                    fiber_wall=0
                    periNum-=2
                    outerFlag=True
                    break
                else:
                    # Length check for inner poly
                    delInnerPolyIdx=[]
                    for innerPlyNum in range(1,len(polygonInnerOuter[polyNum])):
                        ply=polygonInnerOuter[polyNum][innerPlyNum]
                        length=0
                        for indx in range(plastic_wall,plastic_wall+fiber_wall):
                            length+=ply.length
                            ply=ply.buffer(fiberWidth, join_style=1, mitre_limit=15.0)
                        
                        if(length<float(variables.config['fiber']['FIBER_CUT_OFFSET'])):
                            innerOffset=(-fiberWidth+pathSpace)/2 if(periNum==plastic_wall) else pathSpace-fiberWidth
                            innerPolySmall[polyNum].append(polygonInnerOuter[polyNum][innerPlyNum].buffer(innerOffset, join_style=1, mitre_limit=15.0))
                            delInnerPolyIdx.append(innerPlyNum)
                    delInnerPolyIdx.reverse()
                    for idx in delInnerPolyIdx:
                        del polygonInnerOuter[polyNum][idx]
            if outerFlag:
                continue
            
            for periIndx in range(plastic_wall,plastic_wall+fiber_wall):
                if periIndx not in innerSmallPolyPeriCoords.keys():
                    innerSmallPolyPeriCoords[periIndx]=[]
                for polyNumi in range(len(innerPolySmall)):
                    for polyNumj in range(len(innerPolySmall[polyNumi])):
                        if(periIndx==plastic_wall):
                            innerSmallPolyPeriCoords[periIndx].append(list(innerPolySmall[polyNumi][polyNumj].exterior.coords))
                        else:
                            innerPolySmall[polyNumi][polyNumj]=innerPolySmall[polyNumi][polyNumj].buffer(pathSpace,join_style=1, mitre_limit=15.0)
                            innerSmallPolyPeriCoords[periIndx].append(list(innerPolySmall[polyNumi][polyNumj].coords))
            #("II",innerSmallPolyPeriCoords)
            fiberOuterPerimeterCoords,fiberInnerPerimeterCoords,polygonInnerOuterAfterFiber=fiberPeriCoordsGeneration(polygonInnerOuter,plastic_wall,fiber_wall)
            periNum=plastic_wall+fiber_wall-1
        else:
            # Last fiber wall
            if(periNum==plastic_wall+fiber_wall):
                finalPolygon=[]
                #print("TTT",polygonInnerOuterAfterFiber)
                for polyNumi in range(len(polygonInnerOuterAfterFiber)):
                    polygonInnerOuterAfterFiber[polyNumi][0]=polygonInnerOuterAfterFiber[polyNumi][0].buffer(-1*((pathSpace+fiberWidth)/2), join_style=1, mitre_limit=15.0)
                    polyDiff=polygonInnerOuterAfterFiber[polyNumi][0]
                    for innerPlyNum in range(1,len(polygonInnerOuterAfterFiber[polyNumi])):
                        polygonInnerOuterAfterFiber[polyNumi][innerPlyNum]=polygonInnerOuterAfterFiber[polyNumi][innerPlyNum].buffer(((pathSpace+fiberWidth)/2), join_style=1, mitre_limit=15.0)
                        polyDiff=polyDiff.difference(polygonInnerOuterAfterFiber[polyNumi][innerPlyNum])  
                    
                    for polyNum1 in range(len(innerPolySmall)):
                        for polyNum2 in range(len(innerPolySmall[polyNum1])):
                            innerPolySmall[polyNum1][polyNum2]=innerPolySmall[polyNum1][polyNum2].buffer(pathSpace, join_style=1, mitre_limit=15.0)
                            polyDiff=polyDiff.difference(innerPolySmall[polyNum1][polyNum2]) 
                            polygonInnerOuterAfterFiber[polyNumi].append(innerPolySmall[polyNum1][polyNum2])
                            #count+=1
                    finalPolygon.append(polyDiff.simplify(float(variables.config['infill']['SLICING_RESOLUTION']), preserve_topology=False))
                #print("TT",polygonInnerOuterAfterFiber)
            # Further plastic walls after fiber
            else:
                for polyNumi in range(len(polygonInnerOuterAfterFiber)):
                    polygonInnerOuterAfterFiber[polyNumi][0]=polygonInnerOuterAfterFiber[polyNumi][0].buffer(-1*pathSpace, join_style=1, mitre_limit=15.0)
                    for idx in range(1,len(polygonInnerOuterAfterFiber[polyNumi])):
                        polygonInnerOuterAfterFiber[polyNumi][idx]=polygonInnerOuterAfterFiber[polyNumi][idx].buffer(pathSpace, join_style=1, mitre_limit=15.0)
                finalPolygon=polygonDictToPoly(polygonInnerOuterAfterFiber)

        innerPerimeterCoords, outerPerimeterCoords, innerPerimeterArea, outerWallWipeIndices = processWallCoords(periNum, plastic_wall, fiber_wall, finalPolygon, innerPerimeterCoords, innerPerimeterArea, innerPeriCentroidList, outerPerimeterCoords, outerWallWipeIndices)
        periNum+=1
    # Offset the infill polygon (outer/inner walls) as per overlap_factor
    infillPolygons = []
    for poly in finalPolygon:
        if(not poly.is_empty):
            try:
                for ply in poly:
                    polyDiff=Polygon(ply.exterior).buffer(INFILL_WALL_OVERLAP*pathSpace, join_style=1, mitre_limit=15.0)
                    for innerPoly in ply.interiors:
                        polyDiff=polyDiff.difference(Polygon(innerPoly).buffer(-INFILL_WALL_OVERLAP*pathSpace, join_style=1, mitre_limit=15.0))
                    infillPolygons.append(polyDiff)
            except:
                polyDiff=Polygon(poly.exterior).buffer(INFILL_WALL_OVERLAP*pathSpace, join_style=1, mitre_limit=15.0)
                for innerPoly in poly.interiors:
                    polyDiff=polyDiff.difference(Polygon(innerPoly).buffer(-INFILL_WALL_OVERLAP*pathSpace, join_style=1, mitre_limit=15.0))
                infillPolygons.append(polyDiff)
        else:
            infillPolygons.append(Polygon())

    finalInfillPolygons=[]
    for num in range(len(infillPolygons)):
        try:
            for ply in infillPolygons[num]:
                finalInfillPolygons.append(ply)
        except:
            finalInfillPolygons.append(infillPolygons[num])
            pass
    
    for key in innerPerimeterCoords:
        for coordList in innerPerimeterCoords[key]:
            innerPerimeterCoordsList.append(coordList)
    #print("Inner ",innerSmallPolyPeriCoords)
    for key in innerSmallPolyPeriCoords:
        #print(key,innerSmallPolyPeriCoords[key])
        for coordList in innerSmallPolyPeriCoords[key]:
            #print(coordList.xy)
            innerPerimeterCoordsList.append(coordList)
    

    # Inner perimeters sorted by centroid distance
    """ for key in innerPerimeterCoords:
        innerPerimeterCoords[key]=np.array(innerPerimeterCoords[key],dtype=object)
        indexSort=np.argsort(innerPerimeterArea[key])
        indexSort=indexSort.tolist()
        innerPerimeterCoords[key]=innerPerimeterCoords[key][np.argsort(innerPerimeterArea[key])] """
    outerPerimeterCoords = adjustZSeam(outerPerimeterCoords)
    innerPerimeterCoordsList = adjustZSeam(innerPerimeterCoordsList)

    # Wipe for outerPerimeter
    if(float(variables.config['printParameters']['OUTER_WALL_WIPE_DIST'])>0 ):
        for idx in outerWallWipeIndices:
            polyInfill=outerPerimeterCoords[idx]
            travelCoords=outerPerimeterCoords[idx]
            if isinstance(polyInfill[0], list):
                for idx, pointList in enumerate(polyInfill):
                    xPoint, yPoint, index = wipeOuterWall(pointList,float(variables.config['printParameters']['OUTER_WALL_WIPE_DIST']))
                    if(index!=-1):
                        travelCoords[idx].insert(index, (xPoint, yPoint))
                        polyInfill[idx].extend(travelCoords[1:index+1])
                    outerPerimeterCoords[idx] = polyInfill
            else:
                xPoint, yPoint, index = wipeOuterWall(polyInfill,float(variables.config['printParameters']['OUTER_WALL_WIPE_DIST']))
                if(index!=-1):
                    travelCoords.insert(index, (xPoint, yPoint))
                    polyInfill.extend(travelCoords[1:index+1])
                outerPerimeterCoords[idx] = polyInfill
    
    if(variables.config['fiber'].getboolean('USE_FIBER')==True and tempIsInfillLayer):
        return finalInfillPolygons,outerPerimeterCoords,innerPerimeterCoordsList,fiberOuterPerimeterCoords,fiberInnerPerimeterCoords, wallProcessShouldExit
    
    return finalInfillPolygons, outerPerimeterCoords, innerPerimeterCoordsList, wallProcessShouldExit









# function called inside PERIMETERGENERATION

def checkPerimeterCoalesce(outerPolygon, plastic_wall, fiber_wall, fiberWidth, pathSpace):
    """
    Return valid fiber wall count on outer perimeter based on MultiPolygon condition.

    At times, offsetting outer perimeter inwards creates multi-polygon. 
    This is not valid for fiber walls, as fiber walls are required to be continuous.
    Hence valid fiber wall count can be decreased down to where the wall stays continuous.
    """
    # Create polygon for each item of polygonInnerOuter
    # Each polygon can have holes
    validFiberWalls = 0
    firstFiberOffset = (plastic_wall-1)*(pathSpace) + pathSpace/2 + fiberWidth/2
    for idx in range(plastic_wall, plastic_wall+fiber_wall):
        offsetFiberPoly = outerPolygon[0].buffer(-1*(firstFiberOffset+(idx-plastic_wall)*fiberWidth), join_style=1, mitre_limit=15.0)
        # If offsetting the perimeters inwards creates a multi-polygon, fiber cannot be laid on that wall count
        if offsetFiberPoly.geom_type=='MultiPolygon':
            break
        validFiberWalls+=1
    # Reset fiber and plastic wall count based on that
    return validFiberWalls


# FUNCTION CALLED INSIDE PERIMETER GENERATION

def getPerimeterOffsetList(pathSpace, perimeterCount, isInfillLayer, plastic_wall, fiber_wall, fiberWidth):
    """
    This method return a list of buffer offsets to be applied to perimeter walls.

    The offsets are based on whether fiber is used or not.
    
    If only plastic is used, first offset is 0, and rest have value = pathSpace.
    If fiber is used, offsets vary depending on value of wall count.

    A constant plastic inner wall is assumed for all perimeters.
    
    This assumes 2 plastic wall and 1 fiber wall input means 
    = 2 x outermost plastic wall + 1 x fiber wall + 1xplastic wall.
    0 plastic wall and 2 fiber wall input means = 2 x fiber wall + 1xplastic wall
    i.e. plastic inner wall will always be there.
    In special cases, inner plastic wall can be more than 1 
    (depending on cut-distance > perimeter length)
    """
    perimeterOffsetList = []
    if(variables.config['fiber'].getboolean('USE_FIBER')==True and isInfillLayer):
        for wall in range(1, perimeterCount+1):
            if(wall==1):
                perimeterOffsetList.append(0)
            elif(wall<=plastic_wall):
                perimeterOffsetList.append(pathSpace)
            elif(wall==plastic_wall+1 or wall == (fiber_wall+plastic_wall+1)):
                perimeterOffsetList.append((pathSpace+fiberWidth)/2)
            elif(wall>plastic_wall+1 and wall < (fiber_wall+plastic_wall+1)):
                perimeterOffsetList.append(fiberWidth)
            elif(wall>(fiber_wall+plastic_wall+1)):
                perimeterOffsetList.append(pathSpace)
    else:
        for wall in range(1, perimeterCount+1):
            perimeterOffsetList.append(pathSpace if wall>1 else 0)
    return perimeterOffsetList




def periPolyGeneration(offset,polygon):    ### 2 new function called PROCESSOUTERPOLYGON and PROCESSINNERPOLYGON
    """
    Create dict of lists of loops defining walls of perimeters based on offset.

    Each element of the dict is a list. The list contains the outer
    loop of the perimeter count as index '0', and further indices
    contains interior loops. 
    Dict keys are the wall count (-1).
    Offset is based on whether wall is plastic or Fiber.
    """
    # Offset outer perimeter
    outerPolygon=processOuterPolygon(offset, polygon[0]) # polygon[0] contains the polygon

    # Offset all inner perimeters
    innerPolygon=processInnerPolygon(offset, polygon[0]) # polygon[0] contains the polygon
    
    
    # Combine outer and inner perimeter for each body as element in a dict = {0: outerPerimeterOuterWall, 1: innerPerimeter1OuterWall, 2: ....}
    polygonInnerOuter={}
    for polyNum1 in range(len(outerPolygon)):
        polygonInnerOuter[polyNum1]=[]
        polygonInnerOuter[polyNum1].append(outerPolygon[polyNum1])
        for polyNum2 in range(len(innerPolygon)):
            if(innerPolygon[polyNum2].intersects(outerPolygon[polyNum1])):
                polygonInnerOuter[polyNum1].append(innerPolygon[polyNum2])
    
    return polygonInnerOuter


# called inside periPolyGeneration
def processOuterPolygon(offset, polygon):
    """
    Process outer walls out of polygons.
    One polygon  = One loop collection (outer+inner).

    In some cases, polygon can be MultiPolygon as well
    """
    outerPolygon=[]
    outerPolyList = [] 
    if(polygon.geom_type=='MultiPolygon'):
        for poly in polygon:
            outerPolyList.append(Polygon(list(poly.exterior.coords)))
    else:
        outerPolyList.append(Polygon(list(polygon.exterior.coords)))

    for ply in outerPolyList: #For multipolygon handling
        outerPolygon.append(ply.buffer(-1*offset, join_style=1, mitre_limit=15.0))

    return outerPolygon

# called inside periPolyGeneration
# Another fucntion called inside processInnerPolygon i,e Getinteriorlist
def processInnerPolygon(offset, polygon):
    """
    Process inner walls out of polygons. One polygon  = One loop collection (outer+inner).

    In some cases, polygon can be MultiPolygon as well
    """
    innerPoly=[]
    if polygon.geom_type=='MultiPolygon':
        for poly in polygon:
            innerPoly.extend(getInteriorsList(poly, offset))
    else:
        innerPoly.extend(getInteriorsList(polygon, offset))
    try:
        innerPolyUnion=innerPoly
        for innerPolyNum in range(1,len(innerPoly)):
            innerPolyUnion=innerPolyUnion.union(innerPoly[innerPolyNum]) # Union of all the innerpolygon
    except:
        pass
    innerPolygon=[]
    try:
        for ply in innerPolyUnion: #If union returns multipolygon
            innerPolygon.append(ply)
    except:

        try:
            innerPolygon.append(innerPolyUnion)
        except:
            pass
    return innerPolygon


#function called inside processInnerPolygon
def getInteriorsList(polygon, offset):
    innerPoly = []
    for innerPly in polygon.interiors:
        if innerPly.geom_type!="Polygon":
            innerPly=Polygon(innerPly.coords)
        innerPoly.append(innerPly.buffer(1*offset, join_style=1, mitre_limit=15.0))
    return innerPoly



def polygonDictToPoly(polygonInnerOuter):
    """
    Convert the polygon dictionary to polygon with exterior and interiors.

    Each element of 'polygonInnerOuter' contains index '0' as exterior, 
    and rest of the coordinates as interior of a polygon.
    This is converted into a valid polygon.
    This polygon will truly represent the infill polygon for later calculations.
    """
    finalPolygon=[]
    for polyNum in range(len(polygonInnerOuter)):
        polyDiff=polygonInnerOuter[polyNum][0]
        for innerPlyNum in range(1,len(polygonInnerOuter[polyNum])):
            polyDiff=polyDiff.difference(polygonInnerOuter[polyNum][innerPlyNum])  
        finalPolygon.append(polyDiff.simplify(float(variables.config['infill']['SLICING_RESOLUTION']), preserve_topology=False))
    return finalPolygon





def fiberPeriCoordsGeneration(polygonInnerOuter,plastic_wall,fiber_wall):
    fiberOuterPerimeterCoords=[]
    fiberInnerPerimeterCoords=[]
    fiberOuterPeriCoords={}
    fiberOuterPeriCoords[0]=[]
    fiberInnerPeriCords={}
    innerCentroidList=[]
    finalPolyDict={}
    fiberWidth=float(variables.config['fiber']['FIBER_EXTRUSION_WIDTH'])
    for periNum in range(plastic_wall,plastic_wall+fiber_wall):
        finalPolyDict[periNum]=[]
        if(periNum==plastic_wall):
            finalPolygon=polygonDictToPoly(polygonInnerOuter)
            finalPolyDict[periNum]=finalPolygon
        else:
            for polyNumi in range(len(polygonInnerOuter)):
                polygonInnerOuter[polyNumi][0]=polygonInnerOuter[polyNumi][0].buffer(-1*fiberWidth, join_style=1, mitre_limit=15.0)
                for idx in range(1,len(polygonInnerOuter[polyNumi])):
                    polygonInnerOuter[polyNumi][idx]=polygonInnerOuter[polyNumi][idx].buffer(fiberWidth, join_style=1, mitre_limit=15.0)
            finalPolygon=polygonDictToPoly(polygonInnerOuter)
            finalPolyDict[periNum]=finalPolygon

    try:
        for poly in finalPolyDict[plastic_wall][0].interiors:
            if(poly.geom_type!='Polygon'):
                poly=Polygon(poly.coords)
            pointList=list(poly.centroid.coords)
            x_point=pointList[0][0]
            y_point=pointList[0][1]
            fiberInnerPeriCords[str(x_point)+" "+str(y_point)]=[]
            innerCentroidList.append(pointList)
    except:
        print("Fiber printing might break")
        for ply in finalPolyDict[plastic_wall][0]:
            for poly in ply.interiors:
                if(poly.geom_type!='Polygon'):
                    poly=Polygon(poly.coords)
                pointList=list(poly.centroid.coords)
                x_point=pointList[0][0]
                y_point=pointList[0][1]
                fiberInnerPeriCords[str(x_point)+" "+str(y_point)]=[]
                innerCentroidList.append(pointList)
    
    x_end=0 # Used to make cutline for continuous perimeter
    y_end=-500
    
    for periNum in range(plastic_wall,plastic_wall+fiber_wall):
        try:
            cutLine=LineString([Point(x_end,y_end),finalPolyDict[periNum][0].centroid])
            lineString=LineString(finalPolyDict[periNum][0].exterior.coords)
            splitLines=split(lineString,cutLine)
            for lineIdx in range(1,len(splitLines)):
                fiberOuterPeriCoords[0].extend(splitLines[lineIdx].coords)
            fiberOuterPeriCoords[0].extend(splitLines[0].coords)
            count=0
            for poly in finalPolyDict[periNum][0].interiors:
                min=99999
                nearestCentroid=finalPolyDict[periNum][0].centroid
                for centr in innerCentroidList:
                    if(min>Point(centr).distance(poly.centroid)):
                        min=Point(centr).distance(poly.centroid)
                        nearestCentroid=str(centr[0][0])+" "+str(centr[0][1])
                if(poly.geom_type!='Polygon'):
                    poly=Polygon(poly.coords)
                innerCutLine=LineString([Point(x_end,y_end),poly.centroid])
                innerLineStr=LineString(poly.exterior.coords)
                innerSplitLines=split(innerLineStr,innerCutLine)
                for lineIdx in range(1,len(innerSplitLines)):
                    fiberInnerPeriCords[nearestCentroid].extend(innerSplitLines[lineIdx].coords)
                fiberInnerPeriCords[nearestCentroid].extend(innerSplitLines[0].coords)
                count+=1
        except:
            pass
    fiberOuterPerimeterCoords=fiberOuterPeriCoords[0]
    for periCoords in fiberInnerPeriCords:
        fiberInnerPerimeterCoords.append(fiberInnerPeriCords[periCoords])
    
    return fiberOuterPerimeterCoords,fiberInnerPerimeterCoords,polygonInnerOuter




def processWallCoords(periNum, plastic_wall, fiber_wall, finalPolygon, innerPerimeterCoords, innerPerimeterArea, innerPeriCentroidList, outerPerimeterCoords, outerWallWipeIndices):
    if periNum<plastic_wall or periNum>=plastic_wall+fiber_wall:
        if(periNum==0):
            for polyNum in range(len(finalPolygon)):
                try:
                    for innerPly in finalPolygon[polyNum].interiors:
                        pointList=list(innerPly.centroid.coords)
                        x_point=pointList[0][0]
                        y_point=pointList[0][1]
                        innerPerimeterCoords[str(x_point)+" "+str(y_point)]=[]
                        innerPerimeterArea[str(x_point)+" "+str(y_point)]=[]
                        innerPeriCentroidList.append(pointList)
                except:
                    for poly in finalPolygon[polyNum]:
                        for innerPly in poly.interiors:
                            pointList=list(innerPly.centroid.coords)
                            x_point=pointList[0][0]
                            y_point=pointList[0][1]
                            innerPerimeterCoords[str(x_point)+" "+str(y_point)]=[]
                            innerPerimeterArea[str(x_point)+" "+str(y_point)]=[]
                            innerPeriCentroidList.append(pointList)
        # Saving outer and inner perimeter curve coordinates in respective lists
        for polyNum in range(len(finalPolygon)):
            try:
                polyInfill=list(finalPolygon[polyNum].exterior.coords)
                # Add outer wall wipe index for later processing
                if(periNum==0 and float(variables.config['printParameters']['OUTER_WALL_WIPE_DIST'])>0 ):
                    outerWallWipeIndices.append(len(outerPerimeterCoords))
                outerPerimeterCoords.append(polyInfill)
                for innerPly in finalPolygon[polyNum].interiors:
                    ply=innerPly
                    if(ply.geom_type!='Polygon'):
                        ply=Polygon(ply.coords)
                    tempFlag=False
                    for coord in innerPeriCentroidList:
                        if(ply.contains(Point(coord))):
                            x_point=coord[0][0]
                            y_point=coord[0][1]
                            innerPerimeterCoords[str(x_point)+" "+str(y_point)].append(list(innerPly.coords))
                            innerPerimeterArea[str(x_point)+" "+str(y_point)].append(innerPly.area)
                            tempFlag=True
                            break        
            except:
                for poly in finalPolygon[polyNum]:
                    polyInfill=list(poly.exterior.coords)
                    # Add outer wall wipe index for later processing
                    if(periNum==0 and float(variables.config['printParameters']['OUTER_WALL_WIPE_DIST'])>0 ):
                        outerWallWipeIndices.append(len(outerPerimeterCoords))
                    outerPerimeterCoords.append(polyInfill)
                    for innerPly in poly.interiors:
                        ply=innerPly
                        if(ply.geom_type!='Polygon'):
                            ply=Polygon(ply.coords)
                        tempFlag=False
                        for coord in innerPeriCentroidList:
                            if(ply.contains(Point(coord))):
                                x_point=coord[0][0]
                                y_point=coord[0][1]
                                innerPerimeterCoords[str(x_point)+" "+str(y_point)].append(list(innerPly.coords))
                                innerPerimeterArea[str(x_point)+" "+str(y_point)].append(innerPly.area)
                                tempFlag=True
                                break
    
    return innerPerimeterCoords, outerPerimeterCoords, innerPerimeterArea, outerWallWipeIndices





def adjustZSeam(coords):
    """
    Adjust Z-seam of all walls in one direction across layers

    Connect centroid of each wall polygon to the external point.
    Cut the wall polygon with the line thus formed.
    Start from the cut portion, going in one direction in the loop, add all sections to a loop.
    This will align start point of all walls.
    """
    if len(coords) is None or len(coords)==0:
        return coords
    
    newCoords = []
    # If the children are lists
    if len(coords) and isinstance(coords[0], list):
        for coordList in coords:
            if len(coordList):
                element =[]
                polygon = Polygon(coordList)
                polygonLineString = LineString(coordList)
                x_end, y_end, x_start, y_start  = getSeamDirection(polygon.centroid)
                # start_point_x = polygon.centroid.x
                # start_point_y =  polygon.centroid.y
                # Check if the centroid falls inside the polygon
                # if not polygon.centroid.within(polygon):
                # plot.plotPoints([(x_end,y_end),(x_start, y_start)], False, False)
                # plot.plotPoints(coordList,False)
                    # reverse = didOvershoot(x_end, y_end, polygon.centroid.x, polygon.centroid.y, coordList)
                    # start_point_x, start_point_y = extrapolateCentroid(coordList, x_end, y_end, reverse)
                cutLine=LineString([Point(x_end,y_end),Point(x_start,y_start)]) # Line connecting two far end points passing through centroid
                intrsc_points = cutLine.intersection(polygonLineString) # the cutline (can) cut the polygon at multiple points
                nearest_pt = intrsc_points[0]
                if len(intrsc_points)>1:
                    nearest_pt = getNearestPoint(intrsc_points, Point(x_end,y_end)) # Get nearest point to actual seam
                cutLine2 = getCutLine(nearest_pt, Point(x_end,y_end))
                try: # At times the split method fails randomly
                    splitLines=split(polygonLineString, cutLine2)
                    for lineIdx in range(1,len(splitLines)):
                        element.extend(splitLines[lineIdx].coords)
                    element.extend(splitLines[0].coords)
                except: # If split fails, just return original coordinates
                    element.extend(coordList)
            else:
                element=coords
            newCoords.append(element)
    # If the children are tuples of points
    else:
        polygon = Polygon(coords)
        polygonLineString = LineString(coords)
        cutLine=LineString([Point(x_end,y_end),polygon.centroid])
        splitLines=split(polygonLineString,cutLine)
        for lineIdx in range(1,len(splitLines)):
            newCoords.extend(splitLines[lineIdx].coords)
        newCoords.extend(splitLines[0].coords)
    return newCoords



def wipeOuterWall(wallPoints,cutDist):
    """
    This method tells up to which point coordinate the wipe movement should go.

    Method takes outer wall points, and wipe distance as input.
    Starting from first point, it calculates which point lies beyond the wipe distance.
    Then it calculates the exact point (up to 0.05mm accurancy) up to where wipe should happen.
    Then return the point coordinates and its index in the outer wall points' list.
    This index helps the parent function to insert this returned point in the duplicate 
    of the actual outer wall point list.
    The portion from beginning (index=1) up to this index is cut from the duplicate list.
    Then the actual outer wall point list is extended, adding this cut-portion in the end.
    If the length of the outer wall is less than wipe distance, the last point of the list
    is returned, which essentially means that the nozzle will move over the outer wall for
    one whole turn for the wiping action.
    """
    if(len(wallPoints)<2):
        return 0,0,-1
    dist=0
    prevDist=0
    for pointNum, point in enumerate(wallPoints):
        if pointNum==(len(wallPoints)-1):
            return wallPoints[pointNum][0], wallPoints[pointNum][1], pointNum+1
        xCoord=point[0]
        yCoord=point[1]
        prevDist=dist
        dist+=math.hypot(xCoord-wallPoints[pointNum+1][0],yCoord-wallPoints[pointNum+1][1])
        print(f"dist {dist}")
        if(dist>=cutDist):
            xLow=xCoord
            yLow=yCoord
            xHigh=wallPoints[pointNum+1][0]
            yHigh=wallPoints[pointNum+1][1]
            extraDist=cutDist-prevDist
            while(math.hypot(xHigh-xLow,yHigh-yLow)>0.05):
                xMid=(xHigh+xLow)/2.0
                yMid=(yHigh+yLow)/2.0
                if(math.hypot(xMid-xCoord,yMid-yCoord)<extraDist):
                    xLow=xMid
                    yLow=yMid
                else:
                    xHigh=xMid
                    yHigh=yMid
            return xLow,yLow,pointNum+1
    return 0,0,-1
```

```python
from scipy.signal import find_peaks
def complexPolyZoneLines(xcord,ycord,bounds):
    """ This function returns the list of lines which will perfrom polygon slicing i.e. divide the polygon in segments.
    These segments help optimize the infills and travel paths for all types of infills """
    lineCoords=[]
    # print("yCord",ycord)
    #eWidth=float(config['infill']['eWidth'])
    pathSpace=0.4606194490192345
    y=np.array(ycord) # list to numpy array conversion
    y=np.round(y,2) 
    x=np.array(xcord)
    x=np.round(x,2)
    minx,miny,maxx,maxy=bounds
    xRange=maxx-minx 
    yRange=maxy-miny
    minx-=xRange 
    maxx+=xRange
    peaks1=find_peaks(y) # All the maximas of the curve
    y1=-1*y
    peaks2=find_peaks(y1) # All the minimas of the curve 
    # print("peaks1", peaks1)
    # print("peaks2", peaks2)
    yCordPoint=np.concatenate((y[peaks1[0]],y[peaks2[0]])) # y-coordinate of all the maximas and minimas.
    # print("Y Max : ",yCordPoint)
    #xCordPoint=np.concatenate((x[peaks1[0]],x[peaks2[0]])) # x-coordinate of all the maximas and minimas.
    # Sorting X Coordinates relative to Y Coordinates
    #xCordPoints=xCordPoint[np.argsort(yCordPoint)]         
    yCordPoint.sort()
    yCordRefined=[]
    if(len(yCordPoint)==0): #If yCord is not having any maximas of minimas
        return lineCoords,yCordRefined     # no need of yCordRefined
    yCordPoint=np.unique(yCordPoint)
    
    
    #yCordRefined=yCordPoint
    yCordRefined.append(yCordPoint[0])
    # If two maxima/minima are within an eWidth distance, we should consider only one
    for yPoint in range(1,len(yCordPoint)):
        #print("Value : ",abs((abs(yPoint)-abs(yCordRefined[len(yCordRefined)-1]))))
        if(abs((abs(yCordPoint[yPoint])-abs(yCordRefined[len(yCordRefined)-1])))>=pathSpace): #eWidth
            yCordRefined.append(yCordPoint[yPoint])
    
    length=len(yCordRefined)
    flag=False
    # Adding lines through maxima/minima points, from one end to another
    # Each line is traversed reverse to previous to maintain continuity ?
    for paeakNum in range(length):
        point1=(round(minx,2),round(yCordRefined[paeakNum],2))
        point2=(round(maxx,2),round(yCordRefined[paeakNum],2))
        if(flag):
            #print("A ",point1,point2)
            lineCoords.append(point1)
            lineCoords.append(point2)
            flag=False
        else:
            #print("B ",point2,point1)
            lineCoords.append(point2)
            lineCoords.append(point1)
            flag=True
    # print("lineCoords",lineCoords)
    # print("yCordRefined",yCordRefined)

    # xp=[]
    # yp=[]
    # for line in range(len(lineCoords)):
    #     xp.append(lineCoords[line][0])
    #     yp.append(lineCoords[line][1])
    # pyplot.plot(xp,yp)
    # pyplot.show()

    return lineCoords,yCordRefined
```

```python
def processInnerPolygon(offset, polygon):
    """
    Process inner walls out of polygons. One polygon  = One loop collection (outer+inner).

    In some cases, polygon can be MultiPolygon as well
    """
    innerPoly=[]
    if polygon.geom_type=='MultiPolygon':
        for poly in polygon:
            innerPoly.extend(getInteriorsList(poly, offset))
    else:
        innerPoly.extend(getInteriorsList(polygon, offset))
    try:
        innerPolyUnion=innerPoly
        for innerPolyNum in range(1,len(innerPoly)):
            innerPolyUnion=innerPolyUnion.union(innerPoly[innerPolyNum]) # Union of all the innerpolygon
    except:
        pass
    innerPolygon=[]
    try:
        for ply in innerPolyUnion: #If union returns multipolygon
            innerPolygon.append(ply)
    except:

        try:
            innerPolygon.append(innerPolyUnion)
        except:
            pass
    return innerPolygon
```

```python

def wipeOuterWall(wallPoints,cutDist):
    """
    This method tells up to which point coordinate the wipe movement should go.

    Method takes outer wall points, and wipe distance as input.
    Starting from first point, it calculates which point lies beyond the wipe distance.
    Then it calculates the exact point (up to 0.05mm accurancy) up to where wipe should happen.
    Then return the point coordinates and its index in the outer wall points' list.
    This index helps the parent function to insert this returned point in the duplicate 
    of the actual outer wall point list.
    The portion from beginning (index=1) up to this index is cut from the duplicate list.
    Then the actual outer wall point list is extended, adding this cut-portion in the end.
    If the length of the outer wall is less than wipe distance, the last point of the list
    is returned, which essentially means that the nozzle will move over the outer wall for
    one whole turn for the wiping action.
    """
    if(len(wallPoints)<2):
        return 0,0,-1   # ? why
    dist=0
    prevDist=0
    for pointNum, point in enumerate(wallPoints):
        if pointNum==(len(wallPoints)-1): # if it is the last element of wallpoints list
            return wallPoints[pointNum][0], wallPoints[pointNum][1], pointNum+1
        xCoord=point[0]
        yCoord=point[1]
        prevDist=dist
        dist+=math.hypot(xCoord-wallPoints[pointNum+1][0],yCoord-wallPoints[pointNum+1][1]) # Calculating hypotenous
        print(f"dist {dist}")
        if(dist>=cutDist):
            xLow=xCoord
            yLow=yCoord
            xHigh=wallPoints[pointNum+1][0]
            yHigh=wallPoints[pointNum+1][1]
            extraDist=cutDist-prevDist
            while(math.hypot(xHigh-xLow,yHigh-yLow)>0.05):
                xMid=(xHigh+xLow)/2.0
                yMid=(yHigh+yLow)/2.0
                if(math.hypot(xMid-xCoord,yMid-yCoord)<extraDist):
                    xLow=xMid
                    yLow=yMid
                else:
                    xHigh=xMid
                    yHigh=yMid
            return xLow,yLow,pointNum+1 
    return 0,0,-1   #???
```
